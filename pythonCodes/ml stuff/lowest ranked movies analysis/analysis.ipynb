```python
import pandas as pd
```

```python
movies = pd.read_csv('lowest_ranked_movies_data.csv')
movies.head()
```

```python
movies["review_count"] = movies["review_count"].str.replace("[K]","", regex=True).astype(int)
```

```python
movies
```

```python
movies["duration"] = movies["duration"].str.replace("[min]","", regex=True)
```

```python
movies
```

```python
movies["duration"] = movies["duration"].str.replace("[h]","", regex=True)
```

```python
movies
```

```python
movies["duration"] = movies["duration"].str.replace(" ", "")
movies
```

```python
movies.info()
```

```python
movies['duration'].astype(str)
```

```python
movies
```

```python
movies.info()
```

```python
movies['duration'] = movies['duration'].fillna(0).astype(int)
```

```python
movies
```

```python
movies.info()
```

```python
movies['duration'] = movies['duration'] / 100
movies
```

```python
movies.to_csv('movies.csv', index=False)
```

```python
movies.info()
```

```python
import matplotlib.pyplot as plt
plt.scatter(movies['duration'], movies['rating'], marker='^', color='teal', edgecolors='yellow', linewidths=0.5, alpha=0.75)
plt.xlabel('Duration')
plt.ylabel('Rating')
plt.title('Duration vs Rating')
```

From the above distrbutions we can conclude that the movies with durations on the range of roughly 1 hour to 1.5hours were the worst rated movies
